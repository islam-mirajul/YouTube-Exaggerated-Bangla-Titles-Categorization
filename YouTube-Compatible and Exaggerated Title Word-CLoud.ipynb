```
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
import re
from bs4 import BeautifulSoup
from nltk.tokenize import WordPunctTokenizer
```

```
csv = '/content/drive/MyDrive/youtube project/youtubeRD-csv.csv'
my_df = pd.read_csv(csv)
my_df
```

```
my_df.info()
```

```
import matplotlib.font_manager as fm
neg_tweets = my_df[my_df.Type == 'অতিরঞ্জিত']
neg_string = []
for t in neg_tweets.Title:
    neg_string.append(t)
neg_string = pd.Series(neg_string).str.cat(sep=' ')
from wordcloud import WordCloud
wordcloud = WordCloud(font_path='/content/drive/MyDrive/youtube project/Bangla.ttf',width=800, height=400,max_font_size=150,background_color="black").generate(neg_string)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```

```
import matplotlib.font_manager as fm
pos_tweets = my_df[my_df.Type == 'সামঞ্জস্যপূর্ণ']
pos_string = []
for t in pos_tweets.Title:
    pos_string.append(t)
pos_string = pd.Series(pos_string).str.cat(sep=' ')
wordcloud = WordCloud(font_path='/content/drive/My Drive/gender project/kalpurush.ttf',width=800, height=400,max_font_size=150,background_color="black").generate(pos_string) 
plt.figure(figsize=(12,10)) 
plt.imshow(wordcloud, interpolation="bilinear") 
plt.axis("off") 
plt.show()
```
